```python
import ffmpy
import ffmpeg
import os
```

```python
##Exercici 1
def macroblocks(name,output):
    ff = ffmpy.FFmpeg(
        inputs={''+str(name)+'': None},
        outputs={''+str(output)+'': '-flags2 +export_mvs -vf codecview=mv=pf+bf+bb '})
    ff.cmd
    ff.run()
        
        
##Exercici 2
        #Cut BBB into 1 minute only video
def cut(name,output,dur):
    ff = ffmpy.FFmpeg(
        inputs={''+str(name)+'': None},
        outputs={''+str(output)+'': '-ss 00:00:00 -to 00:00:'+str(dur)+''})
    ff.cmd
    ff.run()
        
        #Export BBB(1min) audio as MP3 stereo track.
def mp3stereo(name,output):
    ff = ffmpy.FFmpeg(
        inputs={''+str(name)+'': None},
        outputs={''+str(output)+'': ' -acodec mp3 -vn '})
    ff.cmd
    ff.run()
        
        #Export BBB(1min) audio in AAC w/ lower bitrate
def aacAudio(name,output):
    ff = ffmpy.FFmpeg(
        inputs={''+str(name)+'': None},
        outputs={''+str(output)+'': '-acodec aac -b:a 384k -vn '})
    ff.cmd
    ff.run()

        #Now package everything in a .mp4 with FFMPEG!!   
def packageAll(name,mp3,aac,output):
    ff = ffmpy.FFmpeg(
        inputs={''+ str(name) +'': None,
                ''+ str(mp3)  +'': None,
                ''+ str(aac)  +'': None },
        outputs={''+str(output)+'': ' -map 0 -map 1 -map 2 -codec copy'})
    ff.cmd
    ff.run()
    
        
##Exercici 3  
def broadcasting(name): 
    cmd = "ffprobe -v error -select_streams a:0 \
           -show_entries stream=codec_name -of  \
            default=noprint_wrappers=1:nokey=1 input.mp4 >> codecs.txt"
    os.system(cmd) #find the codec for the video
        
    f = open('codecs.txt','r+')
    codecs = f.read()
    print ("Code Video: "+(codecs)+"")
        
        # Check which codec it is and therefore 
        #  which standard broadcasting is correct
    if codecs == "aac\n":
        print("Broadcasting standard ISDB")
    elif  codecs == "ac3\n":
        print("Broadcasting standard ATSC")
    elif  codecs == "avs\n":
        print("Broadcasting standard DTMB")
    else: print ("ERROR: it doesn’t fit any broadcasting")
    
    f.truncate(0)

        
##Exercici 4
def subtitles(name,output,sub):
    ff = ffmpy.FFmpeg(
        inputs={''+ str(name) +'': None},
        outputs={''+str(output)+'': '-vf subtitles='+str(sub)+''})
    ff.cmd
    ff.run()        
```
